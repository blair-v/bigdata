```python
import os
import pandas as pd

EMPTY_FIELD = '**empty_field**'

def merge_csv(path, output_file):
    files = os.listdir(path)  # pylint: disable-msg=too-many-locals
    files.sort()

    t_df = None

    for file in files:
        print(file)
        path_file = os.path.join(path, file)

        df = pd.read_csv(path_file)

        if t_df is None:
            t_df = df
        else:
            t_df = t_df.append(df)

    t_df.fillna(0)
    #t_df.to_csv(f'{output_file}', index=False)
    return t_df
```
